```python
import cv2
import matplotlib.pyplot as plt

input_image_path = r"C:\Procesamiento de imagenes\Clases\images\serpiente.jpg"
image = cv2.imread(input_image_path, cv2.IMREAD_COLOR)

if image is None:
    print("Error:Image wasnt able to read.")
else:
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title("Original Image")
    plt.axis('off')
    plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap='gray')
plt.title("Image in Grayscale")
plt.axis('off')
plt.show()
```

```python
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

plt.imshow(blurred_image, cmap='gray')
plt.title("Smoothed Image")
plt.axis('off')
plt.show()
```

```python
edges = cv2.Canny(blurred_image, 50, 150)

plt.imshow(edges, cmap='gray')
plt.title("Edge Detection")
plt.axis('off')
plt.show()
```

```python
inverted_edges = cv2.bitwise_not(edges)

plt.imshow(inverted_edges, cmap='gray')
plt.title("Color Inversion")
plt.axis('off')
plt.show()
```

```python
import numpy as np
kernel = np.ones((2, 2), np.uint8)
thick_edges = cv2.dilate(inverted_edges, kernel, iterations=1)

final_image = cv2.bitwise_and(thick_edges, inverted_edges)

plt.imshow(final_image, cmap='gray')
plt.title("Final Image")
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(final_image, cmap='gray')
plt.title("Coloring image")
plt.axis('off')

plt.show()
```
